# Questions 2:
### Accept two lists of 5 elements each from the user. Convert them to numpy arrays. Concatenate these arrays and print it. Also sort these arrays and print it.

```
a=input("Please enter 5 values separated by , for List1")
lst1=[int(i) for i in a.split(",")]
b=input("Please enter 5 values separated by , for List2")
lst2=[int(i) for i in b.split(",")]
arr1=np.array(lst1)
arr2=np.array(lst2)
concarray=np.concatenate((arr1,arr2))
np.sort(concarray)
```

### Write a code snippet to find the dimensions of a ndarray and its size.

# Questions 4:
### How to convert a 1D array into a 2D array? Demonstrate with the help of a code snippet
### Hint: np.newaxis, np.expand_dims

```
arr1=np.array([2,3,4,1,5,7,8,6,3,4,1,23])
print("1D Array:",arr1.shape)
arr1=np.reshape(arr1,(3,4))
print("Converted it to 2D",arr1.shape)
```

# Questions 5:
### Consider two square numpy arrays. Stack them vertically and horizontally.
### Hint: Use vstack(), hstack()

```
one=np.ones((3,3))
zero=np.zeros((3,3))
print("Vertically Stacking: ",np.vstack((one,zero)))
print("Horizontally Stacking: ",np.hstack((one,zero)))
```
